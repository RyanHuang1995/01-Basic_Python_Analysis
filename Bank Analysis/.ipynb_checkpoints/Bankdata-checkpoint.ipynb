```python
#Dependencie Setting
import csv
import os
budget_data_1_path = "budget_data_1.csv"
budget_data_2_path = "budget_data_2.csv"
```

```python
#Import and Read CSV File
with open (budget_data_1_path,newline="") as budget_data_1_file:
    
    budget_data = csv.reader(budget_data_1_file, delimiter = ",")
    budget_data = list(budget_data)
    month = "whatever"
    month_count = 0
    cumulated_revenue = 0
    cumulated_changes = 0
    max_revenue_change = 0
    min_revenue_change = 0
    previous_month_revenue = int(budget_data[1][1])
    
    #The total number of months included in the dataset
    for row in budget_data[1:]:
        if month != row[0]:
            month = row[0]
            month_count += 1
    
    #The total amount of revenue gained over the entire period
    for row in budget_data[1:]:
        cumulated_revenue += int(row[1])
    
    #The average change in revenue between months over the entire period
    for row in budget_data[2:]:
        previous_month_revenue = int(budget_data[1][1])
        monthly_revenue = int(row[1])
        revenue_changed = monthly_revenue - previous_month_revenue
        previous_month_revenue = int(row[1])
        cumulated_changes += revenue_changed
    total_average_changes = round(cumulated_changes/(month_count-1),0)
    
    #The greatest increase in revenue (date and amount) over the entire period
    for row in budget_data[2:]:
        if int(row[1])-previous_month_revenue > max_revenue_change:
            max_revenue_change = int(row[1]) - previous_month_revenue
            previous_month_revenue = int(row[1])
            target_max_month = row[0]
        else:
            previous_month_revenue = int(row[1])
    
    #The greatest decrease in revenue (date and amount) over the entire period
    for row in budget_data[2:]:
        if int(row[1])-previous_month_revenue < min_revenue_change:
            min_revenue_change = int(row[1]) - previous_month_revenue
            previous_month_revenue = int(row[1])
            target_min_month = row[0]
        else:
            previous_month_revenue = int(row[1])
        
    #CONCLUSION
    output_1 = (
    f"\nFinancial Analysis\n"
    f"-------------------\n"
    f"Total Months: {month_count}\n"
    f"Total Revenue: {cumulated_revenue}\n"
    f"Average Revenue Change: {total_average_changes}\n"
    f"Greatest Increase in Revenue: {target_max_month} {max_revenue_change}\n"
    f"Greatest Decrease in Revenue: {target_min_month} {min_revenue_change}\n")
    print(output_1)
```
